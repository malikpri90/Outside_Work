```python
# Identify the seasons played for each player. 
# Remove seasons where game played <=5 
# Rank seasons for each player - e.g. most recent season with >5 games = 1, then 2 and so on. 
# It will be different depending on player, for example 2021 for Zac Williams will be Season 2 whereas for Jack Steele thats season 3


# Extract year from Match Date
joined_data['year'] = pd.DatetimeIndex(joined_data['match_date']).year

# Games player per year. Filter to only where games >= 5, and remove 2023 as we have no response variable for recent scores
player_summ = joined_data.groupby(['player_name','year']).agg(matches=('match_date', 'count'),year_int=('year', 'mean'))\
    .query('matches >=5' )\
    .query('year != 2023') 

# # Row Number to rank seasons in ascending order. 
# player_summ['season_rank_desc'] = player_summ.groupby('player_name')['year'].rank(method="first", ascending=False)
player_summ['season_rank_asc'] = player_summ.groupby('player_name')['year_int'].rank(method="first", ascending=False)

player_summ
```

## Response variable

```python
# Average AFL fantasy score per season
response_var = joined_data.groupby(['player_name','year']).agg(matches=('match_date', 'count'),year_int=('year', 'mean'),avg_score=('afl_fantasy_score', 'mean'))\
    .query('matches >= 5').query('avg_score > 0')\
    .sort_values(by=['avg_score'], ascending=False)

# Subtract 1 year from AFL score so it matches to factors from last year
response_var['predicted_year'] = response_var['year_int'] - 1

response_var  
```

## Build Player Metrics (Last 1 year, Last 3 Year, Last 5 Year)

```python
# Build 1 Year Player Metrics

# Filter to a players most recent season and join to the full dataset
player_summ_1yr = player_summ
# \.query('season_rank_asc == 1')

model_factors_1yr = joined_data.merge(player_summ_1yr, on =['player_name','year'], how='inner') 

model_factors_agg_1yr = model_factors_1yr.groupby(['player_name', 'year'])\
.agg({'player_height_cm': 'mean',
'player_weight_kg': 'mean',
'kicks': 'mean',
'marks': 'mean',
'handballs': 'mean',
'handballs': 'mean',
'disposals': 'mean',
'effective_disposals': 'mean',
'disposal_efficiency_percentage': 'mean',
'goals': 'mean',
'behinds': 'mean',
'hitouts': 'mean',
'tackles': 'mean',
'rebounds': 'mean',
'inside_fifties': 'mean',
'clearances': 'mean',
'clangers': 'mean',
'free_kicks_for': 'mean',
'free_kicks_against': 'mean',
'brownlow_votes': 'mean',
'contested_possessions': 'mean',
'uncontested_possessions': 'mean',
'contested_marks': 'mean',
'marks_inside_fifty': 'mean',
'one_percenters': 'mean',
'bounces': 'mean',
'goal_assists': 'mean',
'time_on_ground_percentage': 'mean',
'centre_clearances': 'mean',
'stoppage_clearances': 'mean',
'score_involvements': 'mean',
'metres_gained': 'mean',
'turnovers': 'mean',
'intercepts': 'mean',
'tackles_inside_fifty': 'mean',
'contest_def_losses': 'mean',
'contest_def_one_on_ones': 'mean',
'contest_off_one_on_ones': 'mean',
'contest_off_wins': 'mean',
'def_half_pressure_acts': 'mean',
'effective_kicks': 'mean',
'f50_ground_ball_gets': 'mean',
'ground_ball_gets': 'mean',
'hitouts_to_advantage': 'mean',
'hitout_win_percentage': 'mean',
'intercept_marks': 'mean',
'marks_on_lead': 'mean',
'pressure_acts': 'mean',
'rating_points': 'mean',
'ruck_contests': 'mean',
'score_launches': 'mean',
'shots_at_goal': 'mean',
'spoils': 'mean',
})

model_factors_agg_1yr.columns = [str(col) + '_1yr' for col in model_factors_agg_1yr.columns]

model_factors_agg_1yr
```

```python
# Most common position for the past year
# Calculate most common position played over past year
# Consider grouping if needed to remove specificity (e.g. Left and Right indicators probably not useful)


# Games player per year. Filter to only where games >= 5
# position_mapp = joined_data.groupby(['player_name','year']).agg({'match_date': 'count'})\
#     .query('match_date >=5')



joined_data['position_grouped'] = np.select(
    [
     joined_data['player_position'].isin(["FB"]), 
     joined_data['player_position'].isin(["BPL", "BPR"]),
     joined_data['player_position'].isin(["CHB"]), 
     joined_data['player_position'].isin(["HBFL", "HBFR"]),
     joined_data['player_position'].isin(["RK"]), 
     joined_data['player_position'].isin(["C", "RR" ,"R"]),
     joined_data['player_position'].isin(["WL", "WR"]), 
     joined_data['player_position'].isin(["FF"]),
     joined_data['player_position'].isin(["FPL", "FPR"]),
     joined_data['player_position'].isin(["HFFL", "HFFR"]),
     joined_data['player_position'].isin(["CHF"]),
     joined_data['player_position'].isin(["INT", "SUB"])
    ],
    [
      'FB',
      'BP',
      'CHB',
      'HBF',
      'RUCK',
      'C',
      'W',
      'FF',
      'FP',
      'HFF',
      'CHF',
      'Ignore'  
    ], 
    default='Other'
     )

## Test no positions have not been allocated. Result = 0 rows
# joined_data.query('position_grouped in ["Other"]')\
#     .groupby(['player_position']).agg(matches=('match_date', 'count'))


## Rank most recent games for each player, ignoring INT and SUB games
joined_data['recent_games'] = joined_data.query('position_grouped not in ["Ignore"]')\
      .groupby('player_name')['match_id'].rank(method="first", ascending=False)\
      

## Filter to most recent 10 games, find most common position. 10 game approach should balance recency while removing 1-2 game outliers
player_pos_all = joined_data.query('recent_games <= 10')\
                                .groupby(['player_name','position_grouped']).agg(matches=('match_id', 'count'))


# Find most common position but most games over past 10
player_pos_all['rank'] = player_pos_all.groupby('player_name')['matches'].rank(method="first", ascending=False)

player_pos_current = player_pos_all.query('rank == 1')

player_pos_current
```

## Join all tables together for model file

```python
model_file = response_var[['predicted_year', 'avg_score']]\
                .merge(player_pos_current, on=['player_name'], how='outer')\
                .merge(model_factors_agg_1yr, left_on=['player_name', 'predicted_year'], right_on=['player_name', 'year'], how='outer')
                               
model_file
```
